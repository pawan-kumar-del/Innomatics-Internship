```python
# Problem 1
def validate_recharge(): 
    valid_plans = [199, 299, 399, 599] 
    while True: 
        amount = int(input("Enter recharge amount: ")) 
        if amount < 50: 
            print("Amount must be at least ₹50. Try again.") 
        elif amount not in valid_plans: 
            print("Invalid plan. Available plans:", valid_plans)
        else: 
            print("Recharge Successful for ₹", amount)
        break 
validate_recharge()
```

```python
# Problem 2
def check_inventory(products):
    for item, qty in products.items():
        if qty < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")
products = {
    "Pen": 10,
    "Notebook": 25,
    "Pencil": 5,
    "Eraser": 18
}
check_inventory(products)
```

```python
# Problem 3
def check_result(marks):
    total = 0
    for m in marks:
        total += m

    avg = total / len(marks)
    print("Average:", avg)

    if avg >= 50:
        print("Pass")
    else:
        print("Fail")

marks = list(map(int, input("Enter marks separated by space: ").split()))
check_result(marks)
```

```python
# Problem 4
def calculate_fare():
    while True:
        distance = float(input("Enter distance (km): "))
        if distance <= 0:
            print("Invalid distance. Try again.")
            continue
        peak = input("Peak hour? (yes/no): ").lower()
        fare = 50 + (distance * 12)
        if peak == "yes":
            fare += fare * 0.25
        print("Total Fare: ₹", round(fare, 2))
        break

calculate_fare()
```

```python
# Problem 5
def attendance_check(records):
    present = 0
    for r in records:
        if r == 'P':
            present += 1
    percentage = (present / len(records)) * 100
    print("Attendance:", percentage, "%")
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
        
records = input("Enter attendance (P/A without space): ").upper()
print(attendance_check(records))
```

```python
# Problem 6
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    if len(password) < 8:
        return "Weak Password"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
